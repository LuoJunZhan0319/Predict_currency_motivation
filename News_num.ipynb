```python
import pandas as pd
import numpy as np
from glob import glob
```

### aud

```python
aud = pd.read_csv('./Pre_data/news/AUDNews105110_5.csv', index_col = 0)
aud_2016 = aud[aud["date"].str.contains("2016.*")]
aud_2017 = aud[aud["date"].str.contains("2017.*")]
aud_2018 = aud[aud["date"].str.contains("2018.*")]
aud_2019 = aud[aud["date"].str.contains("2019.*")]
aud_2020 = aud[aud["date"].str.contains("2020.*")]
aud_2021 = aud[aud["date"].str.contains("2021.*")]

aud_list = [aud_2016,aud_2017,aud_2018,aud_2019,aud_2020,aud_2021]
```

```python
count = 2016
pd_aud = pd.DataFrame()

for i in aud_list:
    name = "aud_"+str(count)
    dic = {} #資料有的tilte都加進去
    for j in i["source_title"]:
        if j not in dic:
            dic[j] = 1
        elif j in dic:
            dic[j] = dic[j]+1
    my_dict = pd.DataFrame(dic,index=[0])
    pd_aud = pd.concat([pd_aud, my_dict])
    np.save('./result/News_resource/year/aud/'+name+'.npy', dic)
    count+=1
```

```python
pd_aud
```

```python
files = glob('./result/News_resource/year/aud/aud_*.npy')

for i in files:
    new_dict = np.load(i, allow_pickle='TRUE')
    print(new_dict)
```
