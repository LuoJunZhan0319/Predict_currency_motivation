### 正式跑SVM

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from scipy.linalg import svd
from sklearn import svm
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
import itertools
import time
```

# AUD

```python
#AUD
training_score = [] #用來儲存訓練結果
testing_score = [] #用來儲存測試結果
z = 1

#滑動視窗設立變數
ty = 16 #訓練年
tm0 = 0 #訓練月十位數
tm1 = 4 #訓練月個位數

for aud_path in aud_paths:
    start = time.time()
    df = pd.read_csv(aud_path,index_col = 0)
    df = df.reset_index() #要重設索引要不然日期會跑掉
    
    if tm1 >= 10:
        if tm0 == 0:
            tm0 = 1
            if tm1 == 10:
                tm1 = 0
            elif tm1 == 11:
                tm1 = 1
            elif tm1 == 12:
                tm1 = 2
            elif tm1 == 13:
                tm1 = 3
        if tm0 == 1:
            if tm1 >= 3:
                tm0 = 0
                tm1 = 1
                ty += 1
    #滑動視窗遮罩    
    test = '20%d-%d%d-01' %(ty, tm0, tm1) #測試資料的切線
    
    print("===================================")
    print("目前使用測試資料切割日期為20{}年{}{}月".format(ty,tm0,tm1))
    
    #滑動視窗範圍內的TFIDF分數
    clean_body = df['clean_body']
    vectorizer = TfidfVectorizer(sublinear_tf=False, stop_words=None, smooth_idf=True, norm='l2')
    tfidf = vectorizer.fit_transform(clean_body) #vectorizer的output是 (文檔索引，詞索引)  tf-idf 的矩陣沒辦法直接拿拿用
    tfidf = tfidf.todense() #要把結果轉為向量表示，維度等於(文章數，所有文章字詞數)，矩陣內容為tfidf值
    
    #建立TFIDF分數資料表
    df_tfidf = pd.DataFrame(tfidf) #redim_tfidf
    df_tfidf['date'] = df['date']
        

    #切割訓練及測試資料
    x_train = df_tfidf[df_tfidf['date'] < test]
    x_test = df_tfidf[df_tfidf['date'] >= test]
    y_train = df[df['date'] < test]
    y_test = df[df['date'] >= test]
    X_train = x_train.drop(['date'], axis=1)
    X_test = x_test.drop(['date'], axis=1)
    
    
    #模型測試 SVM
    clf=svm.SVC(kernel='linear',C=2,gamma='auto')
    clf.fit(X_train,y_train['result'])
    
    
    #模型評估 SVM
    train_score = clf.score(X_train,y_train['result']) 
    test_score = clf.score(X_test, y_test['result'])
    print("第%d次" %(z))
    print(train_score)
    print(test_score)
    end = time.time()
    print("總處理時長:", end - start,"(sec)")
    print("===================================")
    
    #保留評估分數
    training_score.append(train_score)
    testing_score.append(test_score)
    
    #條件判斷
    if tm0 == 1:
        tm1 = 10 * tm0 + tm1
        tm0 = 0
    #視窗滑動   
    tm1 += 1
    
    #計數器
    z += 1
    
    #清空資料
    df = None
```
