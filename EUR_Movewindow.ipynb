#### 儲存滑動視窗資料

```python
import time

staryear = 105
endyear = 108
allmonth = (endyear-staryear+1)*12   #所有的月份資料的總數 -> 60
winsize = 12    #滑動月份大小 (訓練加測試3:1) 訓練75% : 測試25%

training_score = [] #用來儲存訓練結果
testing_score = [] #用來儲存測試結果

#滑動視窗設立變數
sy = 16 #起始年
ey = 17 #結束年
ty = 16 #訓練年
sm0 = 0 #起始月十位數
sm1 = 1 #起始月個位數
em0 = 0 #結束月十位數
em1 = 1 #結束月個位數
tm0 = 1 #訓練月十位數
tm1 = 2 #訓練月個位數

#滑動視窗開始
for z in range(allmonth-winsize):
    
    start = time.time()
    
    #條件判斷(這判斷之累QQQQ)
    if sm1 >= 10:
        if sm0 == 0:
            sm0 = 1
            if sm1 == 10:
                sm1 = 0
            elif sm1 == 11:
                sm1 = 1
            elif sm1 == 12:
                sm1 = 2
            elif sm1 == 13:
                sm1 = 3
        if sm0 == 1:
            if sm1 >= 3:
                sm0 = 0
                sm1 = 1
                sy += 1
                
    if em1 >= 10:
        if em0 == 0:
            em0 = 1
            if em1 == 10:
                em1 = 0
            elif em1 == 11:
                em1 = 1
            elif em1 == 12:
                em1 = 2
            elif em1 == 13:
                em1 = 3
        if em0 == 1:
            if em1 >= 3:
                em0 = 0
                em1 = 1
                ey += 1
                
    if tm1 >= 10:
        if tm0 == 0:
            tm0 = 1
            if tm1 == 10:
                tm1 = 0
            elif tm1 == 11:
                tm1 = 1
            elif tm1 == 12:
                tm1 = 2
            elif tm1 == 13:
                tm1 = 3
        if tm0 == 1:
            if tm1 >= 3:
                tm0 = 0
                tm1 = 1
                ty += 1
        
    #滑動視窗遮罩    
    start = '20%d-%d%d-01' %(sy, sm0, sm1)
    end = '20%d-%d%d-01' %(ey, em0, em1)
    test = '20%d-%d%d-01' %(ty, tm0, tm1)
    
    mask1 = df['date'] >= start
    mask2 = df['date'] < end
    
    #取出滑動視窗範圍資料
    win_df = df[(mask1 & mask2)]
    win_df = win_df.reset_index(drop=True)
    
    #每個滑動視窗各自儲存
    win_df.to_csv("./Pre_data/move_window/EUR_12/"+str(z)+".csv",index = 0)
    
    #條件判斷
    if sm0 == 1:
        sm1 = 10 * sm0 + sm1
        sm0 = 0
    if em0 == 1:
        em1 = 10 * em0 + em1
        em0 = 0
    if tm0 == 1:
        tm1 = 10 * tm0 + tm1
        tm0 = 0
    
    #視窗滑動
    em1 += 1
    sm1 += 1
    tm1 += 1
    print("滑動視窗"+str(z)+"大小:{}".format(win_df.shape))
    print('=============================')
```
