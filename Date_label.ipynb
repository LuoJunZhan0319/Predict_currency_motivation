```python
def getdistributed(sub1):
    up = 0
    down = 0
    balance = 0
    
    up_percent=0
    down_percent=0
    balance_percent = 0

    for i in sub1:
        if i == 0:
            down += 1
        elif i == 1:
            up += 1
        elif i == 2:
            up += 1
    if (up != 0 or down != 0):
        up_percent = (up/(down+up+balance))*100
        down_percent = (down/(down+up+balance))*100
        balance_percent = (balance/(down+up+balance))*100

    return [up_percent,down_percent,balance_percent]
```

```python
def get_set1_cutsize(sub1):
    star="2016-01-01"
    train="2020-01-01"
    validation="2020-04-01"
    test="2020-07-01"
    a= sub1[sub1["date"]<train].shape[0]
    b= sub1[((sub1["date"]>=train) & (sub1["date"]<validation))].shape[0]
    c= sub1[((sub1["date"]<test) & (sub1["date"]>=validation))].shape[0]
    
    print("trainset size:",a)
    print("validationset size:",b)
    print("testset size:",c)
    
    return a,(a+b),(a+b+c)
```

```python
def get_set2_cutsize(sub1):
    star="2016-04-01"
    train="2020-04-01"
    validation="2020-07-01"
    test="2020-10-01"
    a= sub1[sub1["date"]<train].shape[0]
    b= sub1[((sub1["date"]>=train) & (sub1["date"]<validation))].shape[0]
    c= sub1[((sub1["date"]<test) & (sub1["date"]>=validation))].shape[0]
    print("trainset size:",a)
    print("validationset size:",b)
    print("testset size:",c)
    
    return a,(a+b),(a+b+c)
```

```python
def get_set3_cutsize(sub1):
    star="2016-07-01"
    train="2020-07-01"
    validation="2020-10-01"
    test="2021-01-01"
    a= sub1[sub1["date"]<train].shape[0]
    b= sub1[((sub1["date"]>=train) & (sub1["date"]<validation))].shape[0]
    c= sub1[((sub1["date"]<test) & (sub1["date"]>=validation))].shape[0]
    print("trainset size:",a)
    print("validationset size:",b)
    print("testset size:",c)
    
    return a,(a+b),(a+b+c)
```
