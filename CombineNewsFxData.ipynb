```python
import pandas as pd 
import numpy as np 
```

### 合併AUD的檔案

```python
df = pd.read_csv('./Pre_data/news/AUDNews105109.csv', index_col = 0)
fx = pd.read_csv('./Pre_data/forex/AUDUSD_rate_105_109.csv', index_col = 0)
```

```python
df
```

```python
fx
```

```python
fx_df = pd.DataFrame()
fx_df["date"] = fx["date"]
fx_df["binary"] = fx["binary"]
fx_df["ternary_1"] = fx["ternary_1"]
fx_df["ternary_2"] = fx["ternary_2"]
fx_df
```

```python
#以新聞為主去做合併的動作
aud_fx = pd.merge(df,fx_df,how='left')
```

```python
aud_fx
```

```python
#用某一個特定值取代所有 nan
aud = aud_fx.fillna("NAN")
```

```python
#將當日沒有匯率價格走勢影響的新聞(比如六日)，我們都以之後最近一天的匯率最為比較值。
#針對"binary"欄位
for i in range(aud.shape[0]):
    if aud['binary'][i] == 'NAN':
#        print(i)
        for j in range(aud.shape[0]):
            j = j+i+1
            if aud['binary'][j] != 'NAN':
                aud['binary'][i] = aud['binary'][j]
                break

#將當日沒有匯率價格走勢影響的新聞(比如六日)，我們都以之後最近一天的匯率最為比較值。
#針對"ternary_1"欄位
for i in range(aud.shape[0]):
    if aud['ternary_1'][i] == 'NAN':
#        print(i)
        for j in range(aud.shape[0]):
            j = j+i+1
            if aud['ternary_1'][j] != 'NAN':
                aud['ternary_1'][i] = aud['ternary_1'][j]
                break

#將當日沒有匯率價格走勢影響的新聞(比如六日)，我們都以之後最近一天的匯率最為比較值。
#針對"ternary_2"欄位
for i in range(aud.shape[0]):
    if aud['ternary_2'][i] == 'NAN':
#        print(i)
        for j in range(aud.shape[0]):
            j = j+i+1
            if aud['ternary_2'][j] != 'NAN':
                aud['ternary_2'][i] = aud['ternary_2'][j]
                break
```

```python
aud
```

```python
aud_no_wrong_date = aud[aud["date"].str.contains("[0-9]*-[0-9]*-[09]*")]
```

```python
aud_no_wrong_date
```

```python
aud_no_wrong_date.to_csv('./Pre_data/conbine_data/AUDUSD105_109.csv')
```

#### 合併完標簽數!!!!!!!

```python
def label_num(df):
    up = 0
    down = 0
    b = 0

    for i in df:
        if i == 0:
            down += 1
        elif i == 1:
            up += 1
        elif i == 2:
            b += 1

    print("上漲: ",up)
    print("下跌: ",down)
    print("平盤: ",b,"\n") 
```

```python
print("三分類無取範圍")
label_num(aud_no_wrong_date["binary"])
print("三分類，取千分之一範圍")
label_num(aud_no_wrong_date["ternary_1"])
print("三分類，取千分之二範圍")
label_num(aud_no_wrong_date["ternary_2"])
```
